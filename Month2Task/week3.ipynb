```python
# 📦 Import required libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 📊 Step 1: Create custom dataset
# Features: Size of house in square feet
# Target: Price in lakhs
data = {
    'Size (sqft)': [600, 800, 1000, 1200, 1500, 1800, 2000],
    'Price (lakhs)': [30, 38, 45, 52, 65, 75, 85]
}

df = pd.DataFrame(data)

# 🧠 Step 2: Prepare X (features) and y (target)
X = df[['Size (sqft)']]   # 2D array
y = df['Price (lakhs)']   # 1D array

# 🔧 Step 3: Create and train the model
model = LinearRegression()
model.fit(X, y)

# 📈 Step 4: Make predictions
predicted = model.predict(X)

# 📉 Step 5: Visualize
plt.scatter(X, y, color='blue', label='Actual Price')
plt.plot(X, predicted, color='red', label='Predicted Line')
plt.title('House Price Prediction')
plt.xlabel('Size (sqft)')
plt.ylabel('Price (lakhs)')
plt.legend()
plt.grid(True)
plt.show()

# 📋 Step 6: Model details
print("Model Coefficient (slope):", model.coef_[0])
print("Model Intercept:", model.intercept_)
```
