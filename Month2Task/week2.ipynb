```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random data from a normal distribution
data = np.random.normal(loc=50, scale=10, size=1000)

# Plot the distribution
sns.histplot(data, kde=True, color='skyblue')
plt.title("Normal Distribution (Mean = 50, SD = 10)")
plt.show()
```

```python
# Generate random data from a Poisson distribution
data_poisson = np.random.poisson(lam=5, size=1000)

# Plot the distribution
sns.histplot(data_poisson, kde=False, color='lightgreen', bins=15)
plt.title("Poisson Distribution ( = 5)")
plt.show()
```

```python
from scipy.stats import ttest_ind

# Example data: Sales for Strategy A and Strategy B
strategy_A = np.random.normal(200, 20, 100)
strategy_B = np.random.normal(210, 25, 100)

# Perform Independent T-test
t_stat, p_value = ttest_ind(strategy_A, strategy_B)

print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")
```

```python
from scipy.stats import chi2_contingency

# Example contingency table: Customer preference by region
data = [[30, 10],
        [20, 40]]

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

print(f"Chi-Square Statistic: {chi2:.2f}")
print(f"P-Value: {p:.4f}")
```

```python
import scipy.stats as stats

# Example: Confidence interval for Strategy A sales
sample_mean = np.mean(strategy_A)
sample_std = np.std(strategy_A, ddof=1)
n = len(strategy_A)

# 95% Confidence Interval
confidence = 0.95
h = stats.t.ppf((1 + confidence) / 2, n - 1) * (sample_std / np.sqrt(n))
ci_lower = sample_mean - h
ci_upper = sample_mean + h

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Normal, Poisson, Binomial distributions
normal = np.random.normal(0, 1, 1000)
poisson = np.random.poisson(3, 1000)
binomial = np.random.binomial(n=10, p=0.5, size=1000)

plt.figure(figsize=(12, 4))
sns.kdeplot(normal, label="Normal", fill=True)
sns.kdeplot(poisson, label="Poisson", fill=True)
sns.kdeplot(binomial, label="Binomial", fill=True)
plt.legend()
plt.title("Comparison of Probability Distributions")
plt.show()
```

```python
from scipy.stats import ttest_1samp

# Test if the mean of sales is different from 205
sales = np.random.normal(200, 20, 100)

t_stat, p_val = ttest_1samp(sales, 205)

print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_val:.4f}")
```

```python
from scipy.stats import ttest_rel

# Sales before and after a marketing campaign
before = np.random.normal(100, 10, 30)
after = before + np.random.normal(5, 5, 30)

t_stat, p_val = ttest_rel(before, after)

print(f"Paired T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_val:.4f}")
```

```python
from scipy.stats import shapiro

# Test if data is normally distributed
data = np.random.normal(0, 1, 100)

stat, p = shapiro(data)

print(f"Shapiro-Wilk Test Statistic: {stat:.3f}")
print(f"P-Value: {p:.3f}")
```

```python
from scipy.stats import chisquare

# Example: Observed vs Expected frequencies
observed = [50, 30, 20]
expected = [33, 33, 34]

chi_stat, p_val = chisquare(observed, f_exp=expected)

print(f"Chi-Squared Statistic: {chi_stat:.2f}")
print(f"P-Value: {p_val:.4f}")
```
