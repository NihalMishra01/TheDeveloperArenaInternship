```python
# üì¶ Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# üîç Load Titanic dataset from Seaborn
df = sns.load_dataset('titanic')

# ‚úÖ Display first 5 rows
df.head()
```

```python
# ‚ÑπÔ∏è Basic information about the dataset
df.info()
```

```python
# üìä Descriptive statistics for numeric columns
df.describe()
```

```python
# ‚ùì Percentage of missing values in each column
(df.isnull().mean() * 100).round(2)
```

```python
# üîç Value counts of key categorical features
print("Survived:\n", df['survived'].value_counts(), "\n")
print("Pclass:\n", df['pclass'].value_counts(), "\n")
print("Sex:\n", df['sex'].value_counts(), "\n")
print("Embarked:\n", df['embarked'].value_counts(), "\n")
```

```python
# üìà Age distribution plot
plt.figure(figsize=(8, 4))
sns.histplot(df['age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title("Distribution of Age")
plt.show()
```

```python
# üì¶ Boxplot to check outliers in 'fare'
plt.figure(figsize=(8, 2))
sns.boxplot(x='fare', data=df, color='lightgreen')
plt.title("Boxplot of Fare")
plt.show()
```

```python
# üîç Passenger survival across different classes
plt.figure(figsize=(6,4))
sns.countplot(x='pclass', hue='survived', data=df, palette='Set2')
plt.title("Survival by Passenger Class")
plt.show()
```

```python
# üî• Heatmap of numeric feature correlations
plt.figure(figsize=(6,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# üîó Pairwise relationships between numeric variables
sns.pairplot(df[['age', 'fare', 'survived', 'pclass']], hue='survived', palette='husl')
plt.show()
```
