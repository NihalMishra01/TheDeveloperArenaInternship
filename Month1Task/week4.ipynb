```python
# 1. Importing Libraries for Visualization
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# For better plots in Jupyter Notebook
%matplotlib inline
```

```python
# 2. Creating Sample Dataset for Visualization
np.random.seed(42)  # For reproducibility

# Create a DataFrame with numeric and categorical features
data = pd.DataFrame({
    'Age': np.random.randint(20, 60, 100),
    'Salary': np.random.normal(50000, 15000, 100).astype(int),
    'Department': np.random.choice(['HR', 'IT', 'Finance', 'Marketing'], 100),
    'Experience': np.random.randint(1, 15, 100),
    'Satisfaction_Score': np.random.uniform(1, 10, 100)
})

print(data.head())
```

```python
# 3. Basic Line Plot using Matplotlib
plt.figure(figsize=(8, 4))
plt.plot(data['Age'], data['Salary'], marker='o', linestyle='-', color='b')
plt.title('Salary vs Age')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
# 4. Scatter Plot using Matplotlib
plt.figure(figsize=(8, 5))
plt.scatter(data['Experience'], data['Salary'], c='green', alpha=0.6)
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# 5. Histogram using Matplotlib
plt.figure(figsize=(8, 4))
plt.hist(data['Satisfaction_Score'], bins=15, color='purple', alpha=0.7)
plt.title('Distribution of Satisfaction Scores')
plt.xlabel('Satisfaction Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# 6. Boxplot using Seaborn to Check Salary Distribution by Department
plt.figure(figsize=(8, 5))
sns.boxplot(x='Department', y='Salary', data=data, palette='Set2')
plt.title('Salary Distribution by Department')
plt.show()
```

```python
# 7. Violin Plot using Seaborn for Salary by Department (shows distribution and density)
plt.figure(figsize=(8, 5))
sns.violinplot(x='Department', y='Salary', data=data, palette='pastel')
plt.title('Violin Plot of Salary by Department')
plt.show()
```

```python
# 8. Pairplot using Seaborn for Pairwise Relationships in Data
sns.pairplot(data[['Age', 'Salary', 'Experience', 'Satisfaction_Score']], diag_kind='kde')
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```

```python
# 9. Heatmap of Correlation Matrix
plt.figure(figsize=(8, 6))
corr_matrix = data[['Age', 'Salary', 'Experience', 'Satisfaction_Score']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 10. Bar Plot Using Seaborn: Average Salary per Department
avg_salary = data.groupby('Department')['Salary'].mean().reset_index()

plt.figure(figsize=(7, 4))
sns.barplot(x='Department', y='Salary', data=avg_salary, palette='Blues_d')
plt.title('Average Salary by Department')
plt.ylabel('Average Salary')
plt.show()
```

```python
# 11. Creating a Simple Dashboard with Multiple Plots (Matplotlib Subplots)
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Scatter plot
axs[0, 0].scatter(data['Age'], data['Salary'], color='r', alpha=0.5)
axs[0, 0].set_title('Salary vs Age')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Salary')

# Histogram
axs[0, 1].hist(data['Experience'], bins=10, color='orange', alpha=0.7)
axs[0, 1].set_title('Experience Distribution')
axs[0, 1].set_xlabel('Years of Experience')

# Boxplot
sns.boxplot(x='Department', y='Salary', data=data, ax=axs[1, 0], palette='Set3')
axs[1, 0].set_title('Salary by Department')

# Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='viridis', ax=axs[1, 1], cbar=False)
axs[1, 1].set_title('Correlation Heatmap')

plt.tight_layout()
plt.show()
```
